Finding and removing NA values form the dataset

```python
data2.columns[data2.isna().any()]
#used to find the null (na) value. the output is that the hours column is missing a value, and it s what will be filmg in the net cocde
```

```python
data2.hours=data2.hours.fillna(data2.hours.mean())
#I have filled the missing data value with the mean of the hours
```

```python
data2
```

segregate the data for input and output

# New Section

```python
X=data2.iloc[:, :-1].values #pick only the last three columns
print(X.shape)
X
#The input column
```

```python
Y=data2.iloc[:, -1].values
Y
#pick only the last column
#this will be the output
```

Train the dataset using Linear Regresssion

```python
model=LinearRegression()
model.fit(X,Y)
```

Predict the marks based on an input array

```python

a=[[10.2,18,1]]
b=[[2.8,16,0]]
Resultsa=model.predict(a)
Resultsb=model.predict(b)
print(Resultsa)
print(Resultsb)
```
