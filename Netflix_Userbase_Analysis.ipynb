```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('/content/Netflix Userbase.csv')
data.head()
```

```python
#Bar Plot for Gender Distribution
sns.countplot(x='Gender', data=data)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
#gender distribution
plt.figure(figsize=(8,6))
gender = data.Gender.value_counts()
index = ["Male","Female"]
colors = ["m","c"]
exp = [0.01,0.1]
values= gender.values.tolist()
plt.pie(values,
labels=index,
autopct="%.2f%%",
colors=colors,
explode=exp,
shadow=True,
startangle=80)
plt.title('Gender Distribution')
plt.show()
```

```python
#Bar Plot for Age Distribution
sns.countplot(x='Age', data=data)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age Distribution')
plt.show()
```

```python
#Pie chart for the subcription types
sub=data['Subscription Type']. value_counts()
plt.pie(sub, labels=sub.index, autopct='%1.1f%%', startangle=90)
plt.title('Subscription Types')
plt.show()
```

```python
#percentage of subscription
x = data["Subscription Type"].value_counts(ascending=False)
index = [i for i in x.index]
var = [0,0.02,0.1]
plt.figure(figsize=(8,6))
plt.pie(x.values,
labels=index,
autopct="%.2f%%",
explode=var,
shadow=True,
startangle=90)
plt.show()
```

```python
#Box plot -Monthly revenue by subscription type
sns.boxplot(x='Subscription Type', y='Monthly Revenue', data=data)
plt.title('Monthly Revenue by Subscription Type')
plt.xlabel('Subscription Type')
plt.ylabel('Monthly Revenue')
plt.show()
```

```python
#Subscription revenue every month
sns.barplot(x="Subscription Type",
y="Monthly Revenue",
hue="Device",
data=data,
edgecolor="blue",
lw=1)
plt.legend(loc="upper right")
plt.title("Subscription Revenue Monthly")
plt.show()
```

```python
#Heatmap (correlation matrix)
corr_matrix = data[['Monthly Revenue', 'Age']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
#Bar plot - Country distribution
sns.countplot(y='Country', data=data)
plt.title('Country Distribution')
plt.xlabel('Country')
plt.ylabel('Count')
plt.show()
```

```python
#Box plot - Monthly revenue by gender
sns.boxplot(x='Gender', y='Monthly Revenue', data=data)
plt.title('Monthly Revenue by Gender')
plt.xlabel('Gender')
plt.ylabel('Monthly Revenue')
plt.show()
```

```python
# 8. Violin plot - Age distribution by subscription type
sns.violinplot(x='Subscription Type', y='Age', data=data)
plt.title('Age Distribution by Subscription Type')
plt.xlabel('Subscription Type')
plt.ylabel('Age')
plt.show()
```

```python
# 9. Count plot - Device usage
sns.countplot(x='Device', data=data)
plt.title('Device Usage')
plt.xlabel('Device')
plt.ylabel('Count')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()
```

```python
# 10. Scatter plot - Monthly revenue vs. Plan Duration

sns.scatterplot(x='Plan Duration', y='Monthly Revenue', data=data)
plt.title('Monthly Revenue vs. Plan Duration')
plt.xlabel('Plan Duration')
plt.ylabel('Monthly Revenue')
plt.show()
```

```python
data["Join Date"] = pd.to_datetime(data["Join Date"])

# Now you can extract the month

data["Join Month "] = data["Join Date"].dt.month
```
